# Deconvolution d'une image
Le principe du TP est de deconvoluer une image pour rendre visible la plaque d'immatriculation d'une automobile.

## Code

```python
from __future__ import division
import numpy as np
import scipy
from numpy import *
from scipy import *
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndimage

%matplotlib inline


I = imread("car.png")[:,:,1] # The image is already in black & white, so we can just read one color channel (here red)
# I is a N1 x N2 array such that I[i,j], between 0 and 1, codes the intensity of light at pixel (i,j).
# Careful: i is the vertical index and j is the horizontal one! (check it)
figure()
imshow(I, cmap="Greys_r") # Greys_r to display 0 as black and 1 and white

# useful functions: fft2 (not fft!), ifft2 (not ifft!), conj
figure()
imshow(log10(abs(fft2(I))), cmap="Greys_r") # Greys_r to display 0 as black and 1 and white
colorbar()



def deconvolve_image(I, L=15, mu=0.005):
    height, width = I.shape
    h = np.zeros((height, width))


    for j in range(0, 2*L):
        h[0, j]=1./(2*L)


    g_hat = fft2(I)
    h_hat = fft2(h)

    square_h_hat = h_hat*h_hat

    f_hat = g_hat*np.conjugate(h_hat)/(abs(square_h_hat)+mu)

    figure()
    imshow(np.real(ifft2(f_hat)), cmap="Greys_r")
    show()
    
deconvolve_image(I)
```

## Coefficient L
Nous avons obtenu le coefficient L en mesurant l'talement d'un des coins de la plaque d'immatriculation. En effet, ce qui devrait tre un point sur la voiture s'tale sur 30 pixels sur l'image de la voiture. Ainsi, $L = 15$ semble tout indiqu car chaque pixel est une moyenne des 15 pixels sur la gauche et des 15 pixels sur la droite :
$g[i,j] = \frac{1}{30}\Sigma_{n=-15}^{15} f[i+n,j]$.

Notons l'intrt de tourner l'image au pralable, afin que le mouvement de la plaque soit bien horizontal, ce qui facilite grandement les calculs de $h$.

```python
deconvolve_image(I, L=13)
```

```python
deconvolve_image(I, L=17)
```

Avec les deux cas `L=13` et `L=17`, nous observons des rsultats plus flous. On ne note pas de grandes diffrences entre `L = 14`, `L = 15` et `L=16`. Par consquent `L=15` semble tre un choix pertinent. 


## Coefficient $\mu$


Le coefficient $\mu$ permet de ne pas rendre la dconvolution sensible au bruit (le dnominateur est trs proche de zro de la calcul de $\hat{f}$). Nous avons obtenu de bon rsultat avec $\mu = 0.005$ qui est suffisamment petit pour ne pas fausser la dconvolution mais suffisamment grand pour ne pas rendre l'algortihme sensible au bruit.

```python
deconvolve_image(I, mu=0.01)
```

```python
deconvolve_image(I, mu=0.001)
```

Pour la valeur `mu=0.01`, de nombreuses bandes verticales apparaissent  cause des bruits de l'image. Pour `mu = 0.01`, des erreurs sont faites sur la dconvolution la plaque reste trop floue. Le choix de `mu = 0.005` nous parait tre un bon compromis.

Le filtre que nous crons est un passe-haut car en floutant on ne slectionne que les basses frquences. L'opration inverse correspond donc  un passe-haut, car on va slectionn les hautes frquences (les gradients levs dans l'image).
